```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris(as_frame=True)
X_iris = iris.data[['petal length (cm)', 'petal width (cm)']].values
y_iris = iris.target

model = DecisionTreeClassifier(max_depth=2)

model.fit(X_iris,y_iris)
```

```python
from sklearn.svm import SVC
sv = SVC(kernel='poly')
sv.fit(X_iris,y_iris)
sv.score(X_iris,y_iris)
```

```python
score = model.score(X_iris,y_iris)
print(score)
```

Decision Tree Without Regularization

```python
from sklearn.datasets import make_moons

X_moons, y_moons = make_moons(n_samples=150,noise=0.2,random_state=42)
```

```python
model = DecisionTreeClassifier(random_state=42)
model2 = DecisionTreeClassifier(min_samples_leaf=5, random_state=42)
model.fit(X_moons,y_moons)
model2.fit(X_moons,y_moons)
```

```python
X_moons_test, y_moons_test = make_moons(n_samples=1000, noise=0.2, random_state=42)
```

```python
model.score(X_moons_test,y_moons_test)
```

```python
model2.score(X_moons_test,y_moons_test)
```

Regressor Decision Tree

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor

np.random.seed(42)

X_quad = np.random.rand(200, 1) - 0.5
y_quad = X_quad ** 2 + 0.025 * np.random.randn(200,1)
tree_reg = DecisionTreeRegressor(max_depth=2,random_state=42)
tree_reg.fit(X_quad,y_quad)
```
